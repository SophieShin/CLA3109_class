### 실습에 필요한 라이브러리 import

```python
import matplotlib.pyplot as plt
import networkx as nx
import pandas as pd
```

### network 데이터 불러오기 & 시각화

```python
g = nx.karate_club_graph()

plt.figure(figsize =(10, 10))
nx.draw_networkx(g,)
```

### node, edge, adjacency matrix 확인하기

```python
g.nodes()
```

```python
g.edges()
```

```python
nx.adjacency_matrix(g).todense()
```

### 1. degree centrality 측정하기

```python
deg_centrality = pd.Series(nx.degree_centrality(g))
```

### 2. Eigenvecctor centrality 측정하기

```python
eigen_centrality = pd.Series(nx.eigenvector_centrality(g))
```

### 3. Katz centrality 측정하기

```python
katz_centrality = pd.Series(nx.katz_centrality(g))
```

### 4. Page rank 측정하기

```python
page_rank =  pd.Series(nx.pagerank(g))
```

### 5. Closeness Centrality 측정하기

```python
cls_centrality = pd.Series(nx.closeness_centrality(g))
```

### 6. Betweenness Centrality 측정하기

```python
btwn_centrality = pd.Series(nx.betweenness_centrality(g))
```

### centrality들 간의 correlation 측정하기

```python
pd.concat([deg_centrality, eigen_centrality, katz_centrality, page_rank, cls_centrality, btwn_centrality], axis=1).corr()
```
