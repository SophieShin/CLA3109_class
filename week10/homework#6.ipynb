```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
penguins = sns.load_dataset('penguins')
```

## Q1. penguins 데이터 프레임에서 각 종(species) 별로 flipper_length_mm의 histogram을 그리는 코드를 작성하세요. 단, matplotlib 라이브러리만 이용하세요.

```python
df = penguins
df
```

```python
df.species.unique()
```

```python
fig, ax = plt.subplots()
ax.hist(df[df['species'] == 'Adelie']['flipper_length_mm'], bins=8, linewidth=0.5, label = 'Adelie')
ax.hist(df[df['species'] == 'Gentoo']['flipper_length_mm'], bins=8, linewidth=0.5,label = 'Gentoo')
ax.hist(df[df['species'] == 'Chinstrap']['flipper_length_mm'], bins=8, linewidth=0.5, label = 'Chinstrap')
ax.legend()
plt.show()
```

## Q2. penguins 데이터 프레임에서 펭귄의 종(species)을 X축으로 하고 body_mass_g의 평균을 보여주는 bar plot을 그리는 코드를 작성하세요. 이 때, 성별(sex)에 따라 bar가 stacked되도록 하세요. 단, matplotlib 라이브러리만 이용하세요.

```python
species = df['species'].unique()
species
```

```python
male = [df[(df['sex'] == 'Male') & (df['species'] == species[0])]['body_mass_g'].mean(),\
        df[(df['sex'] == 'Male') & (df['species'] == species[1])]['body_mass_g'].mean(), \
        df[(df['sex'] == 'Male') & (df['species'] == species[2])]['body_mass_g'].mean()]
male  
```

```python
female = [df[(df['sex'] == 'Female') & (df['species'] == species[0])]['body_mass_g'].mean(),\
        df[(df['sex'] == 'Female') & (df['species'] == species[1])]['body_mass_g'].mean(), \
        df[(df['sex'] == 'Female') & (df['species'] == species[2])]['body_mass_g'].mean()]
female 
```

```python
fig, ax = plt.subplots()
ax.bar(species, male, label= 'Male')
ax.bar(species, female, label = 'Female', bottom = male)
plt.xlabel('species')
plt.ylabel('Mean of body mass(g)')
plt.legend()
plt.show()
```
